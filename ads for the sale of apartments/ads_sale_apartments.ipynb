## Предобработка данных

### Определим пропущенные значения

```python
for column in data.columns:
    if data[column].isna().sum() > 0:
        print(column, data[column].isna().sum())
```

**- высота потолков**

*Довольно большой объем данных не содержит информации о высоте потолков, удалять нельзя. Попробуем заменить пропуски медианным значением с привязкой к этажности дома (будем исходить из предположения, что дома с одинаковым кол-вом этажей, скорее всего относятся к одному проекту, соответсвенно, имеют схожие параметры).*

```python
data['ceiling_height'] = data['ceiling_height'].fillna(
    data.groupby('floors_total')['ceiling_height'].transform('median'))
```

**- этажность**

*86 в масштабе всего датафрейма - очень маленький набор данных, да и этажность - не такой уж важный для исследования параметр, удалим их.*

```python
data = data.dropna(subset = ['floors_total'])
```

**- жилая площадь в квадратных метрах (м²)**

*Небольшой объем данных не содержит информации о метраже жилой площади, заменим пропуски на медианное значение с привязкой к округленной общей площади.*

```python
data['living_area'] = data['living_area'].fillna(
    data.groupby(
        data['total_area'].round())['living_area'].transform('median'))
print(f'Осталось пропусков: {data["living_area"].isna().sum()}')
```

**- апартаменты**

*Очень большое количество пропусков. Предположим, пропуски в этом столбце означают, что объект недвижимости не относится к апартаментам, поэтому заменим на False.*

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

**- площадь кухни в квадратных метрах (м²)**

*Небольшой объем данных не содержит информации о метраже кухни, может быть, потому что это студия?*

```python
data['studio'].value_counts()
```

```python
data[(data['studio'] == True) & (data['kitchen_area'].isna())]
```

*У всех студий отсутсвует метраж кухни, это логично, но не объясняет все остальные пропуски.*

```python
data['open_plan'].value_counts()
```

```python
data[(data['open_plan'] == True) & (data['kitchen_area'].isna())]
```

*Свободная планировка объясняет еще 67 пропусков, но остается еще много.*

*Заменим пропуски на медианное значение с привязкой к округленной общей площади.*

```python
data['kitchen_area'] = data['kitchen_area'].fillna(
    data.groupby(
        data['total_area'].round())['kitchen_area'].transform('median'))
```

**- количество балконов**

*Очень большое количество пропусков в количестве балконов. Скорее всего, пропуски в этом столбце означают, что балконов нет, поэтому заменим на 0.*

```python
data['balcony'] = data['balcony'].fillna(0)
```

**- название населённого пункта**

*Данные маленькие, при этом отсутсвует очень важный для анализа показатель - название населенного пункта, заменить нельзя, проще удалить.*

```python
data = data.dropna(subset = ['locality_name'])
```

**- расстояние до ближайшего аэропорта и центра города в метрах (м) и число парков и водоемов в радиусе 3 км**

```python
print(f'Расстояние до ближайшего аэропорта, пропуски:{data["airports_nearest"].isna().sum()}')
print(f'Расстояние до центра города, пропуски:{data["cityCenters_nearest"].isna().sum()}')
print(f'Число парков, пропуски:{data["parks_around3000"].isna().sum()}')
print(f'Число водоемов, пропуски:{data["ponds_around3000"].isna().sum()}')
```

*Пропуски в числе парков в том же количетсве, что и в числе водоемов, и в расстоянии до центра (за исключением 1 записи), также у близкого к этому кол-ва объектов отсутсвует информация о расстоянии до аэропорта. Проверим, одни и те же это объекты или разные.*

```python
data[(data['parks_around3000'].isna()) & (data['ponds_around3000'].isna()) &
     (data['cityCenters_nearest'].isna()) & (data['airports_nearest'].isna())]
```

*Выявили 5500 объектов, по которым отсутствует информация по паркам и водоемам в радиусе 3 км, также отсутсвует информация о расстоянии до центра города и ближайшего аэропорта. Такие данные должны быть получены автоматически на основе картографических данных, это странно, что их нет. Такой большой объем пропусков нельзя заполнить без риска исказить результаты исследования, избавляться от них тоже не стоит, так как эти объекты содержат другие важные для исследования данные, оставим эти пропуски.*

**- расстояние до ближайшего парка (м) и до ближайшего водоема (м)**

```python
print(f'Расстояние до парка, пропуски:{data["parks_nearest"].isna().sum()}')
print(f'Расстояние до водоема, пропуски:{data["ponds_nearest"].isna().sum()}')
```

*Очень много пропусков, лучше оставить их как есть*

**-сколько дней было размещено объявление**

*Пропуски в более чем 10% данных, допустим, пропуски связаны с тем, что объект еще продается, тогда пропуски стоит оставить.*

**- количество комнат**

```python
data[(data['rooms'] == 0)]['rooms'].count()
```

*194 квартиры с 0 комнат, возможно, это студии или свободная планировка?*

```python
data[(data['rooms'] == 0)
     & ((data['studio'] == True) | (data['open_plan'] == True))]
```

*194 объекта, относящихся к студиям или свободной планировке, с 0 количеством комнат. Значит, можно заменить кол-во комнат на 1.*

```python
data.loc[(data.rooms == 0), 'rooms'] = 1
data[(data['rooms'] == 0)
     & ((data['studio'] == True) | (data['open_plan'] == True))]
```

*Квартир с 0 комнат не осталось. Проверим, нет ли кварттир-студий или квартир со свободной планировкой 
больше чем с 1 комнатой, что было бы ошибкой.*

```python
data[(data['rooms'] > 1)
     & ((data['studio'] == True) | (data['open_plan'] == True))]
```

*Заменим занчения на 1.*

```python
data.loc[(data['rooms'] > 1) & ((data['studio'] == True) |
                                (data['open_plan'] == True)), 'rooms'] = 1
```

### Изменение типа данных

```python
data.info()
```

*Заменим тип данных в столбце с датами публикации.*

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],
                                              format='%Y-%m-%dT%H:%M:%S')
```

*Заменим тип данных в кол-ве этажей в доме, кол-ве балконов, парков, водоемов на целочисленный, так как эти данные не могут быть выражены дробным значением.*

```python
data['floors_total'] = data['floors_total'].round().astype('Int64')
data['balcony'] = data['balcony'].round().astype('int64')
data['parks_around3000'] = data['parks_around3000'].round().astype('Int64')
data['ponds_around3000'] = data['ponds_around3000'].round().astype('Int64')
```

*Округлим кол-во дней*

```python
data['days_exposition'] = data['days_exposition'].round()
```

*Округилим данные в столбцах с расстояниями, нам не нужна точность до сантиметра* 

```python
data['airports_nearest'] = data['airports_nearest'].round()
data['cityCenters_nearest'] = data['cityCenters_nearest'].round()
data['parks_nearest'] = data['parks_nearest'].round()
data['ponds_nearest'] = data['ponds_nearest'].round()
```

*Заменим тип данных о последней цене на целочисленный и отобразим цены в тыс. для удобства*

```python
data['last_price'] = data['last_price'] / 1000
data['last_price'] = data['last_price'].astype('int64')
```

*Округлим высоту потолка до 2 знаков после запятой.*

```python
data['ceiling_height'] = data['ceiling_height'].round(2)
```

```python
data.info()
```

*Типы данных успешно заменены*

## Расчёты и добавление результатов в таблицу

*Добавим цену за квадратный метр*

```python
data['price_meter'] = (data['last_price'] /
                       data['total_area']).round().astype('int64')
```

*Добавим день недели, месяц и год публикации объявления*

```python
data['day_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
data['month_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

*Добавим столбец с этажом квартиры (по категориям)*

```python
def floor_categorized(row):
    floor = row['floor']
    floor_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floor_total:
        return 'последний'
    return 'другой'


data['floor_category'] = data.apply(floor_categorized, axis=1)
```

*Добавим соотношение жилой и общей площади.*

```python
data['living_ratio'] = data['living_area'] / data['total_area']
```

*Добавим отношение площади кухни к общей.*

```python
data['kitchen_ratio'] = data['kitchen_area'] / data['total_area']
```

*Все столбцы добавлены успешно*

## Общий вывод

В этом исследовании были изучены данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Задача исследования — установить параметры, по которым можно определять рыночную стоимость объектов недвижимости.

Предоставленные данные были обработаны, где была возможность - заполнили пропуски и заменили тип данных. Также убрали совсем редкие и выбивающиеся значения. Было выявлено довольно большое количество пропусков данных, которые должны заполняться автоматически на основе картографических данных, следует обратить на это внимание и выяснить, почему эти данные не заполняются.

В данные были добавлены:
* цена квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

Были изучены следующие параметры: площадь, цена, число комнат, высота потолков.
Также было изучено время продажи квартир, в ходе исследования выяснили, что обычно время продажи находится в пределах от 1,5 до 7 месяцев.

Выяснили, что на стоимость квартиры влияют:
общая площадь - чем больше, тем дороже;
число комнат - чем больше площадь, тем больше комнат, тем выше цена;
удаленность от центра - чем дальше, тем дешевле; 
этаж, на котором расположена квартира - самые дорогие не на первом и не на последнем.

Также есть небольшая зависимость цены за квадратный метр от дня недели и месяца:

разница 1-2 тысячи рублей, в субботу публикуются самые дешевые объявления для всего города, для центра в пятницу;

разница 1-4 тысячи рублей, для всего города самые "дорогие" месяцы с июля по сентябрь и в январе, для центра это август, сентябрь, январь;

самые 'дешевые' объявления публикуются в апреле-мае и октябре для центра, для всего города 'дешевые' месяцы - апрель и декабрь.
